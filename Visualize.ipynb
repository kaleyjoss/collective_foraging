# Parameter settings

Set the following parameter values to visualize the model. If you want to print the output, check the box and see the folder "data".

**targets**: how many resources/food ($N_R$) to generate in the environment.

**beta**: this parameter controls the distribution of targets. $\beta$ == 1.1 generates an environment with lowest clustering (similar to a uniform distribution). As we increase $\beta$, clustering increases, and there is more overlap between targets.

**agents**: how many agents/searchers/foragers ($N_A$) to generate in the environment. 

**mu**: this parameter controls the individual search strategy of the agents, and corresponds to the power-law exponent in $P(l) \approx l^{-\mu}$ where l is the step-length chosen

**alpha**: this parameter controls the probability of an agent to use social information and follow another agent. The probability to follow another agent $P(s) = exp(- \alpha * d)$, where d is the distance between the agents. 

**runs**: how many times you want to run a particular parameter combination.

**tick_speed**: how fast do you want the visuals to run

the generated output file will be named after the parameter combination and contain a single line for every run. 

the columns in the file generated correspond to - [Ticks completed, $N_R$, $N_A$, $\beta$, radius, $\mu$, $\alpha$, average resources found, average distance moved, average search efficiency, average size of the clusters formed, average duration of the clusters formed, total clusters formed, average number of clusters formed simultaneously  ]
